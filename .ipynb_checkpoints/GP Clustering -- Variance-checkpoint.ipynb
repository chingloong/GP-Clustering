```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
 
#plot the scatter plot
def plotScatter(X,Y,title,c):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(X,Y,c+'.')
    ax.set_title(title)
    plt.show()
    
def plotCluster(X,Y,clusters,title):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    colors = ['r','g','b','c','m','y']
    
    c = [colors[((int)(clusters[i])-1) % len(colors)]  + '.' for i in range (len(clusters))]
    for i in range (len(X)):
        ax.plot(X[i],Y[i],c[i])
    ax.set_title(title)
    plt.show()
    
#init parameters for the model
alpha = 1
```

$$ f( x_i, x_j) = \alpha  exp\{ -\frac{1}{2} \sum_{m=1}^d l_m (x_i^m - x_j^m)^2\} $$ 

```python
#kernel function
def kernelFunction(x1, x2):
    c = 0;
    for i in range(len(x1)):
        c += alpha * (x1[i] - x2[i]) **2

    c = math.exp(- 0.5 * c)
    return c
```

$$ C[i,j] = f(x_i,x_j) $$

```python
#calculate the whole covariance matrix 
def getCovarianceMatrix(data):
    size = len(data[0,:])
    C = [[ 0 for i in range (size)] for j in range (size)]
    for i in range (size):
        for j in range (size):
            C[i][j] = kernelFunction(data[:,i], data[:,j])
    return matrix(C)  
```

$$ k = [f(x,x_1), f(x,x_2),...f(x,x_n)]^T $$

```python
#calculate matrix k for new data point x 
def getMatrix_k(x, data):
    size = len(data[0,:])
    k = [ 0 for i in range (size) ]
    for i in range (size):
        k[i] = kernelFunction(x, data[:,i])
    return matrix(k).T
```

```python
#TODO check how to calculate the variance
#calculate variance of dataset
def getVariance(data):
    return np.var(data)
```

$$invA = (C + \sigma_n^2 * I ) ^ {-1} $$

```python
#calculate inversion of Matrix A
def getInversionMatrix(data, C):
    size = len(data[0,:])
    var = getVariance(data)
    return  (C + var * matrix(np.eye(size)) ).I 
```

$$ \sigma^2(x) = \mathscr{K} - k^TC^{-1}k $$
where:
$$ \mathscr{K} = f(x,x) $$
$$ k = [f(x,x_1), f(x,x_2),...f(x,x_n)]^T $$

```python
#calcalate the variance for a given point x
def varianceFunction(x, data, inv):
    k = getMatrix_k(x,data)
    result = kernelFunction(x,x) - k.T * inv * k
    return result[0,0]
```

# Dataset -- R15

```python
dataset = loadtxt("dataset\R15.txt")
data = array([dataset[:,0], dataset[:,1]])
clusters = dataset[:,2]
size = len(clusters)

X = data[0,:]
Y = data[1,:]

#plotScatter(X,Y,"Scatter Plot", 'g')
plotCluster(X,Y,clusters,"Scatter Plot")
```

## alpha = 0.5

```python
grid_x = [0 + 0.5 * i for i in range (40) ] 
grid_y = [20 - 0.5 * i for i in range (40) ] 
alpha = 0.5
variances = [[0 for i in range (len(grid_x)) ] for j in range (len(grid_y))]

C = getCovarianceMatrix(data)
invA = getInversionMatrix(data, C)

for i in range (len(grid_y)):
    for j in range (len(grid_x)):
        variances[i][j] = varianceFunction([grid_x[j], grid_y[i]], data, invA)
```

```python
heatmap = variances
extent = [grid_x[0], grid_x[-1], grid_y[-1], grid_y[0]]

plt.clf()
plt.imshow(heatmap, extent=extent)
plt.colorbar()
plt.show()
```
