#GP Clustering Test

##1. R15 Dataset

```python
#init
dataset = loadtxt("dataset\R15.txt")
data = array([dataset[:,0], dataset[:,1]])
clusters = dataset[:,2]
size = len(clusters)

X = data[0,:]
Y = data[1,:]
dimension = 2
gpc.l = [2,2]
gpc.plotScatter(X,Y,"Scatter Plot", 'g')
```

```python
#calculate Covariance Matrix
C = gpc.getCovarianceMatrix(data)
invA = gpc.getInversionMatrix(data, C)
```

```python
grid_x = [0 + 0.25 * i for i in range (80) ] 
grid_y = [20 - 0.25 * i for i in range (80) ] 
variances = [[0 for i in range (len(grid_x)) ] for j in range (len(grid_y))]

for i in range (len(grid_y)):
    for j in range (len(grid_x)):
        variances[i][j] = gpc.varianceFunction([grid_x[j], grid_y[i]], data, invA)
        
heatmap = variances
extent = [grid_x[0], grid_x[-1], grid_y[-1], grid_y[0]]

gpc.plotHeatMap(heatmap, extent)
```

```python
#get SPEs
SEPs = array([gpc.getEquilibriumPoint(data[:,i], data, invA, 0.5, 1000) for i in range (size) ])
sepList, sepIndexMap = gpc.reduceSEPs(SEPs, 0.99)
print "size of reduced SEPs: " , len(sepList)
gpc.plotScatter(sepList[:,0],sepList[:,1],"reduced SEPs Scatter Plot" ,"r")
```

```python
#get adjacency matrix
A = gpc.getAdjacencyMatrix(sepList, 0.40, 5, data, invA)
sepsClusters = gpc.getSEPsClusters(A, sepList)
print sepsClusters
gpc.plotCluster(sepList[:,0],sepList[:,1],sepsClusters,"Scatter Plot")
```

```python
#get clusters assignment
results = gpc.getPointClusters(sepsClusters, sepIndexMap)
gpc.plotCluster2(X,Y,results,"Scatter Plot")
```

```python
gpc.plotCluster2(X,Y,clusters,"Scatter Plot")
```
